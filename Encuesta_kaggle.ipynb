**Parte 4: Análisis exploratorio de datos parte 2(EDA)
Para el cuarto avance de tu proyecto final comenzaremos con una introducción a la visualización de datos haciendo uso de la misma librería Pandas, para graficar algunas preguntas que nos darán una idea de algunos temas importantes en en el mundo del data science, como lo son el los lenguajes de programación más usados, qué grados académicos tienen las personas que se dedican al análisis de datos y cuáles son los ingresos promedio anuales reportados.**                                                                                                                                                                              Para realizar todo esto, haremos uso del gráfico de barras con la función pandas.DataFrame.plot.bar que nos ayuda a visualizar conteos de frecuencias..

```python
#Visualizar como esta compuesto educacionalmente nuetsra muestra de encuestados.
education = data['Q4'].value_counts(normalize=True , dropna = False).mul(100).reset_index()
education.plot.bar(x='index', y='Q4', color='#884EA0',fontsize=12)
plt.title('Educational level surveyed', fontsize=16 , color = 'blue')


```

```python
jobrole = data['Q5'].value_counts(normalize=True , dropna = False).mul(100).reset_index()
jobrole.plot.bar(x='index', y='Q5', color='#EB984E',fontsize=12)
plt.title('Job Role surveyed', fontsize=16 , color = 'red')
```

```python
language = data['Q7_Part_1'].value_counts(normalize=True , dropna = False).mul(100).reset_index()
language.plot.bar(x='index', y='Q7_Part_1', color='#148F77',fontsize=12)
plt.title('Most used programming language', fontsize=16 , color = 'blue')
```

```python
# Se grafica el ingreso percibido en India, donde curiosamente, del 1 al 2 rango con mayor participacion , hay demasiada
#diferencia en cuantias ganadas por los encuestados, ademas de caida notoria en al porcion que gana dicha cantidad.
income_earned= data[data['Q3']== 'India']['Q24'].value_counts(normalize=True , dropna = True).mul(100).reset_index()
income_earned.plot.bar(x='index', y='Q24', color='#EC7063',fontsize=12)
plt.title('Income earned in India', fontsize=16 , color = 'red')
```

```python
# Se grafica el ingreso percibido en Colombia, donde curiosamente, la mayoria de registros estan en el rando de 1,0-1,999 USD
#y el siguiente dato representativo esta en el rando de 0-999 USD, el salto de rango a rango es estable, y la distribucion
#entre encuestados es mas normalizada.
income_earned= data[data['Q3'] == 'Colombia']['Q24'].value_counts(normalize=True , dropna = True).mul(100).reset_index()
income_earned.plot.bar(x='index', y='Q24', color='#ABEBC6',fontsize=12)
plt.title('Income earned in Colombia', fontsize=16 , color = 'blue')
```

```python
income_earned= data[data['Q2'] == 'Man']['Q24'].value_counts(normalize=True , dropna = True).mul(100).reset_index()
income_earned.plot.bar(x='index', y='Q24', color='#85C1E9',fontsize=12)
plt.title('Income earned by Man', fontsize=16 , color = 'red')
```

```python
income_earned= data[data['Q2'] == 'Woman']['Q24'].value_counts(normalize=True , dropna = True).mul(100).reset_index()
income_earned.plot.bar(x='index', y='Q24', color='#E8DAEF',fontsize=12)
plt.title('Income earned by Woman', fontsize=16 , color = 'red')
```

Las 2 ultimas graficas visualizan el rango de salario ganado por genero , en donde, claramente se aprecia una brecha enorme entre lo que perciben los hombres Vs lo percibido por las mujeres, si bien es cierto hay bajo% de mujeres que ganan una cuantia considerablemente alta,esta es aun menor que la que representa esta misma categoria para el genero masculino, resaltando los grandes retos y oportunidades de la mujer en roles tecnologicos.
